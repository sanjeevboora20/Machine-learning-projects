#### Outlier Handling

```python
df.describe(percentiles=[0.02,0.03,0.97,0.98]).T
```

```python
print(df[df['cholestrol']>353.98].shape)
print(df[df['cholestrol']<160.08].shape)
```

```python
cols_with_outliers = ['cholestrol','rest_bps','old_peak']
```

```python
def outlier_handle(x):
    x = x.clip(upper = x.quantile(0.98))
    x = x.clip(lower = x.quantile(0.02))
    return x
```

```python
df[cols_with_outliers] = df[cols_with_outliers].apply(outlier_handle)
```

```python
df.columns
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test  = train_test_split(x,y,test_size=0.30,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Formula for Feature Importance

feat_imp = (N_t/ N) * (impurity - (N_t_R/N_t) * right_impurity - (N_t_L/N_t) * left_impurity)

where  <br>
1) N = Total number of observations <br>
2) N_t = Total number of observations in the node under consideration<br>
3) N_t_R = Total number of observations in the Right child node under consideration<br>
4) N_t_L = Total number of observations in the Left child node under consideration<br>
5) impurity = gini or entropy for node under consideration.<br>
6) right_impurity = gini or entropy for right child node under consideration.<br>
7) left_impurity = gini or entropy for left child node under consideration.

```python
fimp_chest_pain_m2 = (211/211)*(0.496 - (115/211)*0.35  - (96/211)*0.395)
print(fimp_chest_pain_m2)
```

```python
fimp_m2_chest_pain = (211/211)*(0.496 - ((115/211)*0.35) - ((96/211)*0.395))
print(fimp_m2_chest_pain)
```

```python
fimp_m2_ca1 = (96/211)*(0.395 - ((48/96)*0.08) - ((48/96)*0.5))
print(fimp_m2_ca1)
fimp_m2_ca2 = (115/211)*(0.35 - ((34/115)*0.5) - ((81/115)*0.198))
print(fimp_m2_ca2)
fimp_m2_ca = fimp_m2_ca1 + fimp_m2_ca2
print(fimp_m2_ca)
```

### Cross Validation

```python
from sklearn.model_selection import cross_val_score,KFold
```

#### Applying Cross Validation on Max Depth

```python
# md = max_depth
md = [4,5,6,7,8,9]
kf = KFold(n_splits=7)
for i in md:
    scores = cross_val_score(DecisionTreeClassifier(max_depth=i),x_train,y_train,
                             cv=kf,scoring='accuracy')
    print(f'Scores at max_depth {i} is {scores}')
    print(f'Avg _Score at max_depth {i} is {scores.mean()}')
```

#### Applying Cross Validation on Min_Samples_Split

```python
# mss = min_samples_split 
mss = [7,9,11,13,15,17,19,20,25]
kf = KFold(n_splits=6)
for i in mss:
    scores = cross_val_score(DecisionTreeClassifier(min_samples_split=i),
                             x_train,y_train,scoring='accuracy',cv=kf)
    print(f'Scores at min_samples_split {i} is {scores}')
    print(f'Avg_Score at min_samples_split {i} is {scores.mean()}')
```

#### Applying CV on max_depth and min_samples_split

```python
md = [5,6,7,8,9]
mss = [8,10,12,14,17,20,23]
kf = KFold(n_splits=5)
for i in md:
    for j in mss:
        scores = cross_val_score(DecisionTreeClassifier(max_depth=i,min_samples_split=j),
                                 x_train,y_train,scoring='accuracy',cv=kf)
        print(f'Scores at max_depth {i} and min_samples_split {j} is {scores}')
        print(f'Avg_Score at Scores at max_depth {i} and min_samples_split {j} is {scores.mean()}')
```
